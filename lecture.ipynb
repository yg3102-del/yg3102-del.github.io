```python
import plotly.graph_objects as go

# Filter China
df_china = df[df["Country name"] == "China"].copy()

df_china["Year"] = pd.to_numeric(df_china["Year"], errors="coerce")

df_china = df_china[(df_china["Year"] >= 2019) & (df_china["Year"] <= 2024)]

# Sort by year
df_china = df_china.sort_values("Year")

# Now extract variables
years = df_china["Year"]
life_expectancy = df_china["Explained by: Healthy life expectancy"]
gdp = df_china["Explained by: Log GDP per capita"]

fig = go.Figure()

# Left Y axis: Healthy life expectancy
fig.add_trace(
    go.Scatter(
        x=years,
        y=life_expectancy,
        mode="lines+markers",
        name="Explained by: Healthy life expectancy",
        yaxis="y1"
    )
)

# Right Y axis
fig.add_trace(
    go.Scatter(
        x=years,
        y=gdp,
        mode="lines+markers",
        name="Log GDP per capita",
        yaxis="y2"
    )
)


fig.update_layout(
    title="China – Healthy Life Expectancy vs GDP (Dual Y Axes)",
    xaxis=dict(title="Year"),

    yaxis=dict(
        title="Healthy life expectancy",
        side="left"
    ),

    yaxis2=dict(
        title="Log GDP per capita",
        overlaying="y",
        side="right"
    ),

    template="plotly_white",
    legend=dict(x=0.1, y=1.2)
)

fig.show()
```
