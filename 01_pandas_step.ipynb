# 第一步：Pandas 方案（已就绪）

- 数据：`data/dataset.csv`
- 数值列：`Count of Students`
- 目标：计算 **平均值、 中位数、 众数**

> 如需换数据，覆盖 `data/dataset.csv` 并把 `numeric_column` 改成新的列名即可。

```python

import pandas as pd
from pathlib import Path

csv_path = Path("data/dataset.csv")
numeric_column = "Count of Students"  # 可改

df = pd.read_csv(csv_path)
df[numeric_column] = pd.to_numeric(
    df[numeric_column].astype(str).str.replace(",", "", regex=False).str.strip(),
    errors="coerce"
)
s = df[numeric_column].dropna()

mean_val = s.mean()
median_val = s.median()
modes = s.mode().tolist()

print("Rows:", len(df))
print("Using numeric column:", numeric_column)
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode(s):", modes[:10], "..." if len(modes) > 10 else "")
```
